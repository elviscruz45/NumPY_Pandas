## Merge

```python
iz=pd.DataFrame({"key":["k0","k1","k2","k3"],
                 "a":["a0","a1","a2","a3"],
                 "b":["b0","b1","b2","b3"],})
```

```python
dr=pd.DataFrame({"key_2":["k0","k1","k2","k3"],
                 "c":["c0","c1","c2","c3"],
                 "d":["d0","d1","d2","d3"],})
```

```python
iz
```

```python
dr
```

```python
iz.merge(dr)
```

```python
iz.merge(dr,on="key")
```

```python
iz.merge(dr,left_on="key",right_on="key_2")
```

```python
iz=pd.DataFrame({"key":["k0","k1","k2","k3"],
                 "a":["a0","a1","a2","a3"],
                 "b":["b0","b1","b2","b3"],})

dr=pd.DataFrame({"key_2":["k0","k1","k2",np.nan],
                 "c":["c0","c1","c2","c3"],
                 "d":["d0","d1","d2","d3"],})
```

```python
iz
```

```python
dr
```

```python
iz.merge(dr,left_on="key",right_on="key_2")
```

```python
iz.merge(dr,left_on="key",right_on="key_2",how="left")
```

```python
iz.merge(dr,left_on="key",right_on="key_2",how="right")
```

```python
iz.merge(dr,left_on="key",right_on="key_2",how="inner")
```

```python
iz.merge(dr,left_on="key",right_on="key_2",how="outer")
```
